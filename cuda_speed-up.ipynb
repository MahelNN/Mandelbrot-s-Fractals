## Fourth approach using Cuda for GPU accelerators


```python
import numpy as np
import matplotlib.pyplot as plt

cuda_data = np.loadtxt('Mandelbrot_Cuda/Data/03speedup_grain.dat')
x = np.array([100, 200, 250, 500, 750, 1000])
#plt.figure(figsize=(15,15))
plt.subplot(111)
plt.plot(x, cuda_data[:,2], 's-', label='Tparallel')
plt.plot(x, cuda_data[:,3], 'd-', label='Tcomm')
plt.plot(x, cuda_data[:,1], 'o-', label='Ttotal')
plt.xlabel('$Number \ of \ iterations$')
plt.ylabel('$Computation \  time$')
plt.title('Cuda Grain size')

plt.legend()
plt.grid()
plt.show()
```

```python
cuda_data2 = np.loadtxt('Mandelbrot_Cuda/Data/03speedup_imsize.dat')
size = 768*1024
x2 = np.array([size, 2*size, 3*size, 4*size, 6*size, 8*size, 16*size])
#plt.figure(figsize=(15,15))
plt.subplot(111)
plt.plot(x2, cuda_data2[:,2], 's-', label='Tparallel')
plt.plot(x2, cuda_data2[:,3], 'd-', label='Tcomm')
plt.plot(x2, cuda_data2[:,1], 'o-', label='Ttotal')
plt.xlabel('$Size \ of \ the \ image$')
plt.ylabel('$Computation \  time$')
plt.title('Cuda Image size')

plt.legend()
plt.grid()
plt.show()
```
