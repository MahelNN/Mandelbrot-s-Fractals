## Third approach using OpenMP for muli-core processors

### Different chunk size


```python
import numpy as np
import matplotlib.pyplot as plt

openmp_data4 = np.loadtxt('Mandelbrot_Openmp/Data/02speedup_mw_chunk_4.dat')
openmp_data12 = np.loadtxt('Mandelbrot_Openmp/Data/02speedup_mw_chunk_12.dat')
openmp_data24 = np.loadtxt('Mandelbrot_Openmp/Data/02speedup_mw_chunk_24.dat')
openmp_data48 = np.loadtxt('Mandelbrot_Openmp/Data/02speedup_mw_chunk_48.dat')
openmp_data64 = np.loadtxt('Mandelbrot_Openmp/Data/02speedup_mw_chunk_64.dat')
openmp_data96 = np.loadtxt('Mandelbrot_Openmp/Data/02speedup_mw_chunk_96.dat')

#plt.figure(figsize=(15,15))
plt.subplot(111)
plt.plot(openmp_data4[:,0], openmp_data4[:,1], '>-', label='chunk=4')
#plt.plot(openmp_data12[:,0], openmp_data12[:,1], 's-', label='chunk=12')
#plt.plot(openmp_data24[:,0], openmp_data24[:,1], 'o-', label='chunk=24')
plt.plot(openmp_data48[:,0], openmp_data48[:,1], 'd-', label='chunk=48')
#plt.plot(openmp_data64[:,0], openmp_data64[:,1], 'p-', label='chunk=64')
plt.plot(openmp_data96[:,0], openmp_data96[:,1], 'x-', label='chunk=96')
plt.plot(openmp_data4[:,0], openmp_data4[:,0], 'k', label='linear')
#plt.xlim(0,25)
plt.ylim(0,20)
plt.xlabel('$Number \ of \ processors$')
plt.ylabel('$Speedup$')
plt.title('Master worker - chunk size')

plt.legend()
plt.grid()
plt.show()
```
